```python
# position, name, age, level, salary
se1 = ["Software Engineer", "Max", 20, "Junior", 5000]
se1 = ["Software Engineer", "Lisa", 25, "Senior", 7000]
d1 = ["Designer", "Philipp"]

#def code(se):
#  print(f"{se[1]} is writing code...")
#code(se2)
#code(d1) will work, but shouldn't be possible

#class
class SoftwareEngineer:
#  pass
  # class attribute
  alias = "keyboar Magician"

  def __init__(self, name, age, level, salary):  #dunder methods already provided for us in python
    # instance attribute
    self.name = name
    self.age = age
    self.level = level
    self.salary = salary

  # instance method
  def code(self):
    print(f"{self.name} is writing code...")

  def code_in_language(self, language):
    print(f"{self.name} is writing code in {language}...")

#  def information(self):
#    information = f"name = {self.name}, age = {self.age}, level = {self.level}"
#    return information
  
  def __str__(self):      #string method--object converted to string--similar to java .tostring()
    information = f"name = {self.name}, age = {self.age}, level = {self.level}"
    return information
 
  def __eq__(self, other):    #equal method-- comparing two objects-- .equals(obj) in java
    return self.name == other.name and self.age == other.age

  @staticmethod   #declarator so method will work on the whole class
  def entry_salary(age):
    if age < 25:
      return 5000
    if age < 30:
      return 7000
    return 9000


#instance
se1 = SoftwareEngineer("Max", 20, "Junior", 5000) 
se2 = SoftwareEngineer("Lisa", 25, "Senior", 7000)
se3 = SoftwareEngineer("Lisa", 25, "Senior", 7000)

se1.code() #don't need to use 'self'
se2.code()
se1.code_in_language("Python")
se2.code_in_language("C++")
#print(se1.information())
#print(se2.information())

print(se1)
print(se2)

print(se2 == se3) #without the eq method, this would return false as it will compare the memory addresses

print(se1.entry_salary(24)) #would not work if static method declaration missing
SoftwareEngineer.entry_salary(27)
```

##Section 3 - Inheritance

```python
# inherit, extend, override

class Employee: #parent class
#  pass

  def __init__(self, name, age, salary):  
    self.name = name
    self.age = age
    self.salary = salary

  def work(self):
    print(f"{self.name} is working...")


class SoftwareEngineer(Employee): #child class-inherits Employee

    def __init__(self, name, age, salary, level): #overriding init
      super().__init__(name, age, salary) #super refers to parent class 
      self.level = level

    def work(self): #overriding
      print(f"{self.name} is coding...")

    def debug(self): #extending
     print(f"{self.name} is debugging...")
     

class Designer(Employee): #child class- inherits Employee

    def work(self): #overriding
      print(f"{self.name} is designing...")

    def draw(self): #extending
     print(f"{self.name} is drawing...")


se = SoftwareEngineer("Max", 25, 6000, "junior")
print(se.name, se.age)
se.work()
print(se.level)

d = Designer("Philipp", 27, 7000)
print(d.name, d.age)
d.work()
d.draw()
```

```python
# Polymorphism
#Gives us a way to use a class exactly like its parent but each child class keeps its own methods.
employees = [SoftwareEngineer("Max", 25, 6000, "Junior"),
             SoftwareEngineer("Lisa", 30, 9000, "Senior"),
             Designer("Philipp", 27, 7000)]

def motivate_employees(employees): #works on all types of Employees
  for employee in employees:
    employee.work()

motivate_employees(employees)
```

###Summary of Section 3
1. inheritance (parent vs. child)
2. inherit, extend, override
3. super().__init__()
4. polymorphism

```python
#Encapsulation: Mechanism of hidding data implementation and restricting the access
#Abstraction: Natural extension of encapsulation. Each object should only expose a high level mechanism for using it.
#should hide internal implenitation details. only reveal operations relevant to the objects.

class SoftwareEngineer:

  def __init__(self, name, age):
    self.name = name
    self.age = age
    self._salary = None #protected attribute
    self.__salary = None #private attribute (inacessable from the outside)
    self.__num_bugs_solved = 0

  def code(self):
    self.__num_bugs_solved += 1

  # getter              getter and setter methods allows for modification on private attibutes
  def get_salary(self):
    return self.__salary

  # setter
#  def set_salary(self, value):
#    # check value, enforce constraints, etc.
#    if value < 1000:
#      self.__salary = 1000
#    if value > 20000:
#      self.__salary = 20000
#    self.__salary = value

  def set_salary(self, base_value):
    self.__salary = self.__calculate_salary(base_value)

  def __calculate_salary(self, base_value):
    if self.__num_bugs_solved < 10:
      return base_value
    if self.__num_bugs_solved < 100:
      return base_value * 2
    return base_value * 3


se = SoftwareEngineer("Max", 25)
print(se.age, se.name)
print(se.age, se.name, se._salary)
##print(se.age, se.name, se.__salary)
se.set_salary(6000)
print(se.get_salary())

for i in range(70):
  se.code()

se.set_salary(5000)
print(se.get_salary())

##se.__calculate_salary(5000)
```
