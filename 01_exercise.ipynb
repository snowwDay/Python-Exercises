```python
#Create a function that takes a string and returns the number of words, assuming that space(s) seperates words.
str = input("Enter words seperated by a space: ")
def words(str):
  l = str.split()
  return len(l)

print(words(str))
  

```

```python
#Create a function that takes a string and returns the number of words, assuming that a comma separates words.
str = input("Enter words seperated by a comma: ")
def words(str):
  l = str.split(",")
  return len(l)

print(words(str))
```
