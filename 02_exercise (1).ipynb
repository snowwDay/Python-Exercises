```python
#Write a function (say "choose_one()") takes two integers and returns one of the followings:
#(1) one if they are identical;
#(2) smaller one if both are even;
#(3) larger one if both are odd; or
#(4) even if one is even and the other is odd.
#NOTE: You can use any built-in functions (e.g., min(), max(), and others) if needed.
def choose_one(a, b):
  if a == b:
    return a
  if a % 2 == 0 and b % 2 == 0 :
    return min(a,b)
  if a % 2 != 0 and b % 2 != 0:
    return max(a,b)
  if a % 2 == 0 and b % 2 != 0:
    return a
  else:
    return b
  
a = int(input("enter a: "))
b = int(input("enter b: "))
choose_one(a,b)
```

```python
#Write a function that returns true if the both words in the two-word string (passed as an actual parameter) end with the same letter.
def does_it_end_with_same(myString):
  sep = myString.split()
  return sep[0][len(sep[0])-1] == sep[1][len(sep[1])-1]

does_it_end_with_same("live life")
```

```python
#Write a function that takes one string and returns both the numbers of lower and upper cases.
#Example: count_number("Sam Houston State University") returns [21, 4].
def count_number(str):
  up = 0
  low = 0
  for i in str:
    if i.isupper():
      up = up + 1
    else:
      low = low + 1
  return [low,up]

count_number('HellO')
```

```python
#Write a function that take a list of integers and return the information of the most frequent integer.
#Assume that there is a single highest frequecy number.
#Example: find_most_frequent([1, 1, 1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5]) returns [4, 5], 
#where 4 is the frequent integer and 5 is its frequency.
def find_most_frequent(myList):
  myList.sort()
  num = myList[0]  
  c = 0
  for i in myList:
    frequency = myList.count(i)
    if(frequency > c):
      c = frequency
      num = i
  return [num,c]
 
find_most_frequent([1,5,5,3,8,8,8])
  
```

```python
#Write a function that take a list of numbers and return sum of them.
def compute_sum(myList):
  sum = 0
  for i in myList:
    sum = sum+i
  return sum

compute_sum([1,2,3,4])
```

```python
#Write a function that take a list of numbers and returns product of them.
def compute_product(myList):
  prod = 1
  for i in myList:
    prod = prod * i
  return prod

compute_product([2,3,4])
```
